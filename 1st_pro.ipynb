```python
import pandas as pd
```

```python
data = pd.read_csv('student_performance.csv')
print(f"Total Students: {len(data)}")
print(f"Total Columns: {data.shape[1]}")
print(f"Total Departments: {data['department'].nunique()}")
```

```python
dept = data.groupby('department').size()
print(dept)
```

```python
print("highest maths score : ",data.max()['math_score'])
print("lowest maths score : ",data.min()['math_score'])
print("Average maths score : ",data.mean(numeric_only=True)['math_score'])
print("Average programming score : ",data.mean(numeric_only=True)['programming_score'])
print("Average Attendance : ",data.mean(numeric_only=True)['attendance_percentage'])
```

```python
print("Gender wise Average score")
print("-"*40)
total_score = data['math_score']+data['programming_score']+data['science_score']+data['english_score']
data['total_score'] = total_score
gender_score = data.groupby('gender')['total_score'].mean().round(2)
print(gender_score)
```

```python
print("Top 5 Performers:")
print("-"*40)

top_performers = data.sort_values(by='total_score', ascending=False).head(5)

for index, row in top_performers.iterrows():
    print(f"Name: {row['name']}, Total Score: {row['total_score']}")
```

```python
print("Students of Lower Attendance")
print("="*50)
low_at = data[data['attendance_percentage']<75][['name','department','attendance_percentage']]
if len(low_at)>0:
  print(low_at.to_string(index=False))
else:
  print("All the Students are Greater than 75 %")
```
